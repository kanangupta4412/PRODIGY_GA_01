```python
from aitextgen import aitextgen
from aitextgen.colab import mount_gdrive, copy_file_from_gdrive
```

### **Loading GPT-2**

```python
ai = aitextgen(tf_gpt2="124M", to_gpu=False)
```

### **Mounting Google Drive**

```python
mount_gdrive()
```

### **Uploading the custom data file**

```python
file_name = "KB.json"
```

```python
copy_file_from_gdrive(file_name)
```

### **Fine-tuning GPT-2**

```python
ai.train(file_name,
         line_by_line=False,
         from_cache=False,
         num_steps=3000,
         generate_every=2000,
         save_every=2000,
         save_gdrive=False,
         learning_rate=1e-3,
         fp16=False,
         batch_size=1,
         )
```

### **Loading the trained model**

```python
ai = aitextgen(model_folder="trained_model", to_gpu=True)
```

### **Generating text from the Trained model**

```python
ai.generate()
```

```python
ai.generate(n=2,
            prompt="How to treat mental health.",
            max_length=200,
            temperature=2.0,
            top_p=0.9)
```
